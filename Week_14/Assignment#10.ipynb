```python
# yelp_class = df[df.stars.isin([1, 5])]
yelp_class = df[(df.stars==5) | (df.stars ==1)]
```

```python
yelp_class
yelp_class.stars.unique() # check only class 1 and 5 
```

```python
X = yelp_class.text.values
y = yelp_class.stars.values
```

```python
corpus = []

for i in range(len(X)):
    text = re.sub('[^a-zA-Z]',' ',X[i]).lower().split()
    text_len.append(len(text))
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    new_text = [ ps.stem(word) for word in text if word not in set(all_stopwords)]
    new_text = ' '.join(new_text)
    corpus.append(new_text)

print(corpus[0])
```

```python
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
y = y
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
clf = MultinomialNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
