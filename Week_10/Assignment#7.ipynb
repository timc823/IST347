```python
import pandas as pd
import numpy as np
!pip install --upgrade mlxtend

from mlxtend.preprocessing import TransactionEncoder 
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
```

```python
df = pd.read_csv('/content/drive/MyDrive/IST347/Week_10/Market_Basket_Optimisation.csv', header=None)
```

```python
df.fillna('none', inplace=True)
df=df.values
```

```python
df
```

```python
# remove space infront 
for i in range(len(df)):
    for j in range(len(df[i])):
        df[i][j] = df[i][j].strip()
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(df)
```

```python
data = pd.DataFrame(te_array, columns= te.columns_).drop('none', axis =1)
```

```python
data.head()
```

```python
frequent_pattern = apriori(data, min_support=0.005, use_colnames=True, max_len=4)
frequent_pattern 
```

```python
res = association_rules(frequent_pattern, metric = 'confidence', min_threshold=0.4)
res1 = res.loc[:,['antecedents','consequents','support','confidence','lift']]
```

```python
sorted_res1 = res1.sort_values(by='lift', ascending=False)
sorted_res1
```

From the result, I can see people are getting different ingredients with the spaghetti. The most popular options are gound beef and shrimp, follow by frozen vegetables, olive oil, tomoatoes and grated cheese. I think what store can do is it can stock these items in a close area so people can get all items they are looking for in a short distance. 

```python
frequent_fpg = fpgrowth(data, min_support= 0.005 , use_colnames=True, max_len=4)
frequent_fpg
```

```python
res_fpg = association_rules(frequent_fpg, metric = 'confidence', min_threshold=0.4)
res2= res_fpg.loc[:,['antecedents','consequents','support','confidence','lift']]
```
