```python
df = pd.read_csv('/content/drive/MyDrive/IST347/Week_10/BigSports Purchase Data.csv', header=None)
```

```python
df.head()
```

```python
df.fillna('none', inplace=True)
```

```python
df = df.values
df
```

```python
from mlxtend.preprocessing import TransactionEncoder 
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(df)
```

```python
data = pd.DataFrame(te_array, columns= te.columns_).drop('none', axis=1)
```

```python
data
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_pattern = apriori(data, min_support=0.2, use_colnames=True)
frequent_pattern 
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_pattern, metric = 'confidence', min_threshold=0.4)
res
```

```python
res1 = res[['antecedents','consequents','support','confidence','lift']]
res1
```

```python
sorted_res1 = res1.sort_values(by='lift', ascending=False)
print(sorted_res1)
```
